```python
image = cv2.imread("rabbit.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image.shape
```

```python
vgg.input_shape
```

```python
image = cv2.resize(image, (224, 224))
plt.imshow(image)
```

```python
image.shape
```

```python
image_ = np.expand_dims(image, axis=0)
```

```python
image_.shape
```

```python
# VGG preprocessing
image_ = tf.keras.applications.vgg16.preprocess_input(image_)
```

```python
image_.shape
```

```python
plt.imshow(image_[0])
```

```python
result = vgg.predict(image_)
result
```

```python
np.argmax(result)
```

```python
tf.keras.applications.vgg16.decode_predictions(result, top=10)
```

```python
dog = cv2.cvtColor(
    cv2.imread("dog-puppy-on-garden-royalty-free-image-1586966191.jpg"),
    cv2.COLOR_BGR2RGB)
dog.shape
```

```python
dog = cv2.resize(dog, (224,224))
dog = np.expand_dims(dog, axis=0)
```

```python
plt.imshow(dog[0])
```

```python
dog = tf.keras.applications.vgg16.preprocess_input(dog)
result = vgg.predict(dog)
tf.keras.applications.vgg16.decode_predictions(result)
```
